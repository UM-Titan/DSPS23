```python
import json,os, cv2, random
import pandas as pd
import numpy as np 
from google.colab.patches import cv2_imshow
import random
```

```python
## Generate unique colors for each class and severity 
def get_colors ():
  colors = {}
  for i in range(1,19):
    colors[i] = (random.randint(1,255), random.randint(1,255),random.randint(1,255))

  return colors

def convert_bbox(cbox, img_w, img_h, names, cimg, rows):
  ccolor = int(cbox[0])+1
  cbox = cbox[1:]
  x11, y11 = int((cbox[0] - cbox[2]/2)*img_w), int((cbox[1] - cbox[3]/2)*img_h)
  x22, y22 = int((cbox[0] + cbox[2]/2)*img_w), int((cbox[1] + cbox[3]/2)*img_h)
  rows.append([cimg, x11,y11,x22,y22,ccolor,names[ccolor]])
  return rows
```

```python
def show_annotations(input_folder,csv):
  names = { 1:'longitudinal_high',2:'longitudinal_low',3:'longitudinal_medium', 
           4:'grass', 5:'patch_high', 6:'manhole_high', 7:'transverse_high',
           8:'transverse_low', 9:'transverse_medium', 10:'diag_high', 
           11:'diag_low', 12:'diag_medium', 13:'alligator_high', 
           14:'alligator_low', 15:'alligator_medium',  16:'block_low', 
           17:'block_high', 18:'block_medium'}
  
  
  font = cv2.FONT_HERSHEY_SIMPLEX
  org = (50, 50); fontScale = 1; color = (255, 0, 0); thickness = 2  

  colors = get_colors ()
  rows = []
  all_files = os.listdir(input_folder)
  random.shuffle(all_files)
  df = pd.read_csv(csv)
  print (df['cls'].unique())
  # print (df.head())
  for cimg in all_files[0:3]:
    cimg_path = os.path.join(input_folder,cimg)
    annotations = df[df['img_name'] == cimg]
    img_file__ = cv2.imread(cimg_path);
    img_h, img_w, _ = img_file__.shape
    img_w, img_h
    print (cimg)
    # print (annotations)
    for data in annotations.values.tolist():
      # print (data)
      cimg,x1,y1,x2,y2=data[0:-2]
      ccolor = data[-2] + '_' + data[-1]
      
      value = [i for i in names if names[i]==ccolor]
      color = colors[value[0]]

      print (cimg,x1,y1,x2,y2,ccolor,color)
      rows.append([cimg,x1,y1,x2,y2,ccolor,color])
      img_file__ = cv2.rectangle(img_file__, (x1, y1), (x2, y2), color, 3)
      img_file__ = cv2.putText(img_file__, ccolor, (x1,y1), font, 
                    fontScale, color, thickness, cv2.LINE_AA)
    cv2_imshow(img_file__)

```

```python
csv_file = 'task2_data/annotations.csv'
show_annotations('task2_data/images',csv_file)
```
